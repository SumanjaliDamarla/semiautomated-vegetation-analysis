```python
#download for different months and 2 areas, then check for processing level and cloud data
STUDY_AREA = "StudyArea.geojson"
FIRST_DATE = date(2018, 11, 1)
LAST_DATE = date(2018, 11, 30)
SATELLITE = "Sentinel-2"
PROCESSING_LEVEL = "Level-2"
```

```python
api = SentinelAPI( USERNAME, PASSWORD, API_SOURCE)
```

```python
footprint = geojson_to_wkt(read_geojson(STUDY_AREA))
```

```python
products = api.query(footprint,
                     date=(FIRST_DATE, LAST_DATE),
                     platformname=SATELLITE
                    )
```

```python
products.keys()
```

```python
products_df = api.to_dataframe(products)
```

```python
products_df
```

```python
product_urls = []
for i in range(products_df.shape[0]):
    if PROCESSING_LEVEL in products_df["processinglevel"][i]:
        product_urls.append(products_df["link_alternative"][i])
```

```python
keys = list(products.keys())
for i in range(products_df.shape[0]):
    if PROCESSING_LEVEL not in products_df["processinglevel"][i]:
        products.pop(keys[i])
```
